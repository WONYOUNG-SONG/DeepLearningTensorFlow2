**Linear Hypothesis**

<div>
    
</div>



| x | y |
|---|---|
| 1 | 1 |
| 2 | 2 |
| 3 | 3 |

Which hypothesis is the most accurate when the regression (data) looks like the table above?

In this case, we can use linear hypothesis and linear equation.

$$
H(x) = Wx + b
$$


Now, we simply calculate the distance between our hypothesis and the data, i.e. the difference in values.

We call it **Cost (Loss) function**.
$$
(H(x) - y)^2
$$

Why do we square this? To make the difference positive and penalize more when the it is large.

In this example, the equation is:

$$
\frac{(H(x_1)-y_1)^2+(H(x_2)-y_2)^2+(H(x_3)-y_3)^2}{3}
$$

If n is number of data,
$$
cost(W, b) = \frac{1}{n}\sum_{i=1}^{n}(H(x_i)-y_i)^2
$$


_Goal: Minimize cost(W, b)_


**Make Linear Regression with TensorFlow2**

<div>
  

```python
import numpy as np
import tensorflow as tf

#use numpy array to make vectors
x_train = np.array([1, 2, 3, 4])
y_train = np.array([0, -1, -2, -3])

# groups a linear stack of layers into a model
tf.model = tf.keras.Sequential()

# units == output shape, input_dim == input shape
# Dense layer automatically handles the creation of weights and biases
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))

# sgd = tf.keras.optimizers.SGD(lr=0.1)  # SGD == standard gradient descendent, lr == learning rate
sgd = tf.keras.optimizers.SGD(learning_rate=0.1) #need to use learning_rate instead of lr
tf.model.compile(loss='mse', optimizer=sgd)  # mse == mean_squared_error, 1/m * sig (y'-y)^2

# prints summary of the model to the terminal
tf.model.summary()

# fit() executes training
# epochs is how many times the algorithm is going to run
tf.model.fit(x_train, y_train, epochs=100)

# print final weight and bias
final_weights, final_biases = tf.model.layers[0].get_weights()
print("Final Weights:", final_weights)
print("Final Biases:", final_biases)

# predict() returns predicted value
y_predict = tf.model.predict(np.array([5, 4]))
print(y_predict)
```

We received the value of -1 for weight and the value of 1 for bias.
So,
$$
H(x) = -1 * x + 1
$$
